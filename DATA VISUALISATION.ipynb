# relational plot

```python
data=pd.read_csv('tips.csv')
```

```python
data.head()
```

```python
snr.relplot(x='total_bill',y='tip',hue='sex',data=data)
```

```python
snr.relplot(x='total_bill',y='tip',hue='smoker',style='sex', data=data)
```

```python
snr.relplot(x='total_bill',y='tip',hue='smoker',style='sex',col='time', data=data)
```

```python
snr.relplot(x='total_bill',y='tip',hue='smoker',style='day', data=data)
```

# cat plot

```python
snr.catplot(x='total_bill',y='day',data=data)
```

```python
#More tips were collected during dinner time or lunch time?
snr.catplot(x='tip',y='time',data=data)
```

```python
# who is giving more tip 

snr.catplot(x='tip',y='sex',data=data)
```

# box plot

```python
snr.catplot(x='day',y='tip',hue='sex',kind='box',data=data)

# MALES ARE GIVING MORE TIP
# RIGHT SKEWED = MEAN>MEDIAN
# LEFT SKEWED = MEDIAN>MEAN

# ON SUNDAY BOTH MALE AND FEMALE BOXPLOT IS LEFT SKEWED i.e MEDIAN>MEAN
# ON SATURDAY FEMALE IS BALANCED AND MALE IS LEFT SKEWED i.e MEDIAN>MEAN
# ON THURSDAY FEMALE AND MALE BOXPLOT IS RIGHT SKEWED i.e MEAN>MEDIAN
# ON FRIDAY FEMALE AND MALE BOXPLOT IS LEFT SKEWED i.e MEDIAN>MEAN

# THE BLACK DOTS ARE OUTLIARS
```
