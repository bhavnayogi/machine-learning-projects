```python
import numpy as np
import pandas as pd
```

```python
#GATHERING DATA
data=pd.read_csv("insurance.csv")
```

```python
data.head()
```

```python
#DATA PREPROCESSING
#one hot encoding
data['sex']=data['sex'].apply({'female': 1, 'male': 0}.get)
```

```python
data.head()
```

```python
data['smoker']=data['smoker'].apply({'yes': 1, 'no': 0}.get)
data['region']=data['region'].apply({'southeast': 1, 'southwest': 0,'northeast': 2,'northwest': 3}.get)
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
data.info()
```

```python
#DIVIDE THE DATA INTO DEPENDENT AND INDEPENDENT VARIABLES
x=data[['age','sex','bmi','children','smoker','region']]
y=data['expenses']
```

```python
x.head()
```

```python
y.head()
```

```python
#SPLITTNG THE DATA INTO TRAINING AND TESTING SET
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3) #20% of the dat used for testing
```

```python
#MLM USING LINEAR REGRESSION ALGORITHM

from sklearn.linear_model import LinearRegression
regression=LinearRegression() #activating the algorithm
```

```python
regression.fit(x_train,y_train) #learning takes place
```

```python
y_pred=regression.predict(x_test)
```

```python
print(y_pred)
```

```python
#CHECKING ACCURACY OF OUR MODEL

regression.score(x,y)
```
